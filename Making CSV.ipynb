```python
import tabula
import pandas as pd
import numpy as np
import os
```

```python
directory = '/Users/taksa/Desktop/Internship/'
new_dir = []
for file in os.listdir(directory):
    if file.endswith(".PDF"):
        new_dir.append(file)
new_dir
```

```python
def pdf_to_csv(pdf):
    df = tabula.read_pdf(pdf, pages='all')
    new = pd.DataFrame()
    pol = df[0].columns[0].split(':')
    new[pol[0]] = ['\''+pol[1]]
    new[df[1].iloc[0,0]] = [df[1].iloc[0,1]]
    new[df[2].iloc[0,0]] = df[2].iloc[0,1]
    new[df[2].iloc[0,2]] = df[2].iloc[0,3]+df[2].iloc[1,3]
    new[df[2].iloc[2,0]] = df[2].iloc[2,1]
    new[df[2].iloc[2,2]] = '\''+df[2].iloc[2,3]
    df[3].fillna(" ", inplace = True)
    df[3].iloc[1,:] = df[3].iloc[1,:] + df[3].iloc[2,:]
    df[3].iloc[3,:] = df[3].iloc[3,:] + df[3].iloc[4,:]
    df[3].iloc[6,:] = df[3].iloc[6,:] + df[3].iloc[7,:]
    df[3].iloc[9,:] = df[3].iloc[9,:] + df[3].iloc[10,:]
    df[3].iloc[11,:] = df[3].iloc[11,:] + df[3].iloc[12,:]
    df[3].iloc[13,:] = df[3].iloc[13,:] + df[3].iloc[14,:]

    df[3].drop([2,4,7,10,12,14], axis=0, inplace=True)
    
    head = df[3].iloc[:,0].values
    value = df[3].iloc[:,1].values

    for i in range(len(head)):
        new[head[i]] = [value[i]]
    new

    head = df[3].iloc[:,2].values
    value = df[3].iloc[:,3].values

    for i in range(len(head)):
        new[head[i]] = [value[i]]
        
    head = df[4].iloc[0,:].values
    value = df[4].iloc[1,:].values

    for i in range(len(head)):
        new[head[i]] = [value[i]]
        
    df[5] = df[5].drop(0,axis=0)

    head = df[5].iloc[:,0].values
    value = df[5].iloc[:,2].values

    for i in range(len(head)):
        if head[i] != value[i]:
            new[head[i]] = [value[i]]

    head = df[5].iloc[:,3].values
    value = df[5].iloc[:,5].values

    for i in range(len(head)):
        if head[i] != value[i]:
            new[head[i]] = [value[i]]
            
    head = df[6].iloc[:,0].values
    value = df[6].iloc[:,2].values

    for i in range(len(head)):
        if head[i] != value[i]:
            new[head[i]] = [value[i]]
    
    lst = df[11].columns
    ad = df[11]
    
    new['GST (Rs)'] = float(lst[2])+float(ad.iloc[0,2])+float(ad.iloc[3,2])+float(ad.iloc[4,2])

    return new
```

```python
df1 = pdf_to_csv(new_dir[0])
df2 = pdf_to_csv(new_dir[1])
df3 = pdf_to_csv(new_dir[2])
df4 = pdf_to_csv(new_dir[3])
df5 = pdf_to_csv(new_dir[4])
```

```python
df1.to_csv('one.csv')
df2.to_csv('two.csv')
df3.to_csv('three.csv')
df4.to_csv('four.csv')
df5.to_csv('five.csv')
```
